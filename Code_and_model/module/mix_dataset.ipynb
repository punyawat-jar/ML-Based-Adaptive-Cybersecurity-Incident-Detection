```python
def label_preprocess(df, label):
    print(label)
    df.loc[df['label'] == "normal", "label"] = 0
    df.loc[df['label'] == label, "label"] = 1
    df.loc[(df['label'] != 0) & (df['label'] != 1), "label"] = 0
    df['label'] = df['label'].astype('int')
    
    df = pd.get_dummies(df, columns=['protocol_type', 'service', 'flag'])
    for col in df.columns:
        if df[col].dtype == 'bool':
            df[col] = df[col].astype(int)
        
    scaler = MinMaxScaler()
    df[df.columns] = scaler.fit_transform(df[df.columns])
    
    return df
```

```python
labels = ori_df.label.value_counts().index.tolist()
labels
```

```python
for i, label in enumerate(labels):
    df_temp = pd.concat([df1, df2])
    
    print(f'Starting {label} {i+1}/{len(labels)}')
    df_temp = label_preprocess(df_temp, label)
    df_temp.to_csv(f".\\{directory}\\{label}.csv", index=False)
```

```python
def preprocess(df):
    
    # df.loc[df['label'] == "normal", "label"] = 0
    # df.loc[df['label'] != 0, "label"] = 1
    # df.loc[(df['label'] != 0) & (df['label'] != 1), "label"] = 0
    # df['label'] = df['label'].astype('int')
    df = pd.get_dummies(df, columns=['protocol_type', 'service', 'flag'])
    for col in df.columns:
        if df[col].dtype == 'bool':
            df[col] = df[col].astype(int)
        
    # Initialize the MinMaxScaler
    scaler = MinMaxScaler()
    
    # Selecting columns to scale (all columns except 'label')
    columns_to_scale = df.columns.drop('label')
    
    # Applying the scaler to these columns
    df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])
    
    return df
```
